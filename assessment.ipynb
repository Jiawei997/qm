```python
key_fields=['host_about','host_response_rate','host_total_listings_count',
            'accommodates','amenities', 'price','minimum_nights','number_of_reviews',
            'review_scores_rating','room_type','host_is_superhost','instant_bookable']

key_fields_dic={'host_about':'review_scores_rating','host_response_rate':'review_scores_rating',
                'host_total_listings_count':'review_scores_rating','accommodates':'review_scores_rating',
               'amenities':'review_scores_rating','price':'review_scores_rating',
               'minimum_nights':'review_scores_rating','number_of_reviews':'review_scores_rating',
               'room_type':'review_scores_rating','host_is_superhost':'review_scores_rating',
               'instant_bookable':'review_scores_rating'}

key_fields_dol={'var':['host_about','host_response_rate','host_total_listings_count',
            'accommodates','amenities', 'price','minimum_nights','number_of_reviews',
            'review_scores_rating','room_type','host_is_superhost','instant_bookable']
}
key_fields_lol=[['review_scores_rating'],
               ['host_about','host_response_rate','host_total_listings_count',
            'accommodates','amenities', 'price','minimum_nights','number_of_reviews',
            'room_type','host_is_superhost','instant_bookable']
               ]
for i in range(0,11):
    #print(key_fields_lol[1][i])
    X = d1[key_fields_lol[1][i]]
    Y = d1[key_fields_lol[0]]
    plt.xlabel(key_fields_lol[1][i])
    plt.ylabel(key_fields_lol[0])

    a = plt.plot(X,Y,'.', label='original data')
    plt.xlim(0,r)
    plt.ylim(b,c)

    slope, intercept, r_value, p_value, std_err = sps.linregress(X, Y) 

    b = plt.plot(X, X*slope + intercept, 'r', label = 'fitted line') # Plot the regression line.

    plt.legend()
    plt.show()

    print("y =", round(slope,3), "x +", round(intercept,3))
    print("Rsq = ", r_value**2)
    print("p-value = ", p_value)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# It turns out it's easier to work with columns with single word names, so...
d1 = d1.rename(columns = {'2018_cases_ratio':'ratio',
                         'media_awareness':'media',
                          'school_awareness':'school',
                         'health_training':'health'})

# This performs a multiple regression of wellbeing against greenery and natural light:
multi_regression_model_1 = smf.ols(formula='ratio ~ media + sub_counselling + school + health + clean_air + clean_environ',data=d1).fit()

# And this gives a big summary of the results:
print(multi_regression_model_1.summary())
```

```python
# Pick out some values of interest and print the regression equation:

beta_0, beta_1, beta_2, beta_3, beta_4, beta_5, beta_6 = multi_regression_model_1.params
rsq                    = multi_regression_model_1.rsquared
pval_0, pval_1, pval_2, pval_3, pval_4, pval_5, pval_6 = multi_regression_model_1.pvalues

print("y =", round(beta_1,3), "x1 +", round(beta_2,3), "x2 +", round(beta_3,3),"x3 +",round(beta_4,3),"x4 +",round(beta_5,3),"x5 +",round(beta_6,3),"x6 +" round(beta_0,3))
print("R^2     = ", rsq)
print("p-value_1 = ", round(pval_1,3))
print("p-value_2 = ", round(pval_2,3))
print("p-value_3 = ", round(pval_3,3))
print("p-value_4 = ", round(pval_4,3))
print("p-value_5 = ", round(pval_5,3))
print("p-value_6 = ", round(pval_6,3))
```

```python
# produce regression plots
fig = plt.figure(figsize=(10,8))
fig = sm.graphics.plot_regress_exog(multi_regression_model_1, 'media', fig=fig)
```
