### simple linear regression

```python
# This is a regression of well-being against greenery:
def sim_regression(x:'str',y:'str',xmax:'int',ymin:'float',ymax:'float'):
    X = d1[x]
    Y = d1[y] 
    plt.xlabel(x)
    plt.ylabel(y)
    a = plt.plot(X,Y,'.', label='original data')
    plt.xlim(0,xmax)
    plt.ylim(ymin,ymax)

    slope, intercept, r_value, p_value, std_err = sps.linregress(X, Y) 

    b = plt.plot(X, X*slope + intercept, 'r', label = 'fitted line') # Plot the regression line.

    plt.legend()
    plt.show()

    print("y =", round(slope,3), "x +", round(intercept,3))
    print("Rsq = ", r_value**2)
    print("p-value = ", p_value)
```

```python
sim_regression('host_about','review_scores_rating',1000,0,100)
sim_regression('host_response_rate','review_scores_rating',100,0,100)
sim_regression('host_is_superhost','review_scores_rating',1,0,100)
sim_regression('host_total_listings_count','review_scores_rating',510,0,100)
sim_regression('room_type','review_scores_rating',5,0,100)
sim_regression('accommodates','review_scores_rating',5,0,100)
sim_regression('amenities','review_scores_rating',5,0,100)
sim_regression('amenities','review_scores_rating',60,0,100)
sim_regression('price','review_scores_rating',1250,0,100)
sim_regression('minimum_nights','review_scores_rating',365,0,100)
sim_regression('number_of_reviews','review_scores_rating',800,0,100)
sim_regression('instant_bookable','review_scores_rating',1.5,0,100)
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
multi_regression_model_1 = smf.ols(formula='''review_scores_rating ~ host_about + host_response_rate + 
host_total_listings_count + accommodates + amenities + price+minimum_nights+number_of_reviews
+room_type+host_is_superhost+instant_bookable''',data=d1).fit()
print(multi_regression_model_1.summary())
```
